```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
data = pd.read_csv('covid19.csv')
N = 25
states = data['states'].values
tot = data['total'].values

ind = np.arange(N)  # the x locations for the groups
width = 0.35      # the width of the bars

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
rects1 = ax.bar(ind, tot, width, color='royalblue')

cured =  data['Cured'].values
rects2 = ax.bar(ind+width/2, cured, width, color='green')

deaths = data['Death'].values
rects3 = ax.bar(ind+width+width/2, deaths, width, color='red')

# add some
ax.set_ylabel('No of people')
ax.set_title('COVID-19 cases in India')
plt.xticks(ind + width /2 , rotation = 'vertical')
ax.set_xticklabels( states )
ax.legend( (rects1[0], rects2[0], rects3[0]), ('Confirmed COVID-19 cases', 'Cured/Discharged', 'Deaths') )
plt.show()
```

# **Visualization for the comparision of daily growth in cases in COVID19 between Italy and India**



```
# daywise data for rise in covid19 cases in different countries
data = pd.read_csv('daily-cases-covid-19.csv')
```

```
# renaming the humungous column name for ease of use
data.rename(columns={'Daily new confirmed cases of COVID-19 (cases)': 'confirmed'}, inplace=True, copy=False)
```

```
# making the values in date column into datetime format for calculations
data['Date'] = pd.to_datetime(data.Date)
```

```
# fetching for dates after 01 Mar, 2020
filtered_data = data[data.Date >= '2020-03-01']
```

```
filtered_data
```

```
# filtering two separate datasets for the two countries, i.e India and Italy
india = filtered_data[filtered_data.Code == 'IND'] 
italy = filtered_data[filtered_data.Code == 'ITA']
```

```
india
```

```
italy
```

```
# plotting to see the diffence between Italy and india COVID19 cases

fig2 = plt.figure(figsize=(20,6))
plt.plot(italy.Date, italy.confirmed)
plt.xlim('2020-03-01', '2020-03-27')
plt.yticks(np.arange(min(italy.confirmed), max(italy.confirmed)+1, 400))
plt.plot(india.Date, india.confirmed)
plt.legend(['Italy', 'India'])
plt.xlabel('Days')
plt.ylabel('No pf people affected')
plt.title('COVID 19 cases growth during March 2020')
plt.show()
```

# **Visualization for daily percentage growth in the two countries**

```
# setting the relative percentage of growth in each of the datasets. (relative to the highest record in the respective set)
india.confirmed = (india.confirmed / india.confirmed.max()) * 100
italy.confirmed = (italy.confirmed / italy.confirmed.max() ) * 100 
```

```
fig2 = plt.figure(figsize=(20,6))
plt.plot(italy.Date, italy.confirmed)
plt.xlim('2020-03-01', '2020-03-27')
plt.plot(india.Date, india.confirmed)
plt.legend(['Italy', 'India'])
plt.xlabel('Days')
plt.ylabel('Relative percentage')
plt.title('COVID 19 cases relative percentage wise growth during March 2020')
plt.show()
```
